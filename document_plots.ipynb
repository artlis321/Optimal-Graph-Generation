```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg

xs = np.linspace(0,1,100)
ys = np.exp(-xs) + xs / np.e

f,ax = plt.subplots(1, 1)
f.set_size_inches(10, 5)
ax.plot(xs,ys)
ax.set_ylim(0.7,1.2)
ax.set_xlabel("Total edge length (cost)")
ax.set_ylabel("Total path length (inefficiency)")
ax.set_xticklabels([])
ax.set_yticklabels([])

newax = f.add_axes([0.65,0.25,0.2,0.2], anchor='NE', zorder=1)
im = plt.imread('complete.png')
newax.imshow(im)
newax.axis('off')

newax = f.add_axes([0.1,0.6,0.2,0.2], anchor='NE', zorder=1)
im = plt.imread('steiner.png')
newax.imshow(im)
newax.axis('off')

newax = f.add_axes([0.35,0.35,0.2,0.2], anchor='NE', zorder=1)
im = plt.imread('question.png')
newax.imshow(im)
newax.axis('off')

plt.show()
```
